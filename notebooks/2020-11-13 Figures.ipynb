```python
import sys, os 
import cv2
import exrex
import matplotlib
%matplotlib inline
# matplotlib.use('tkAgg')
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import random
import re
from termcolor import colored
import traceback
```

```python
def make_image(R, C, size=15*3, indices=[], filename='../outputs/output.png'):
	img = np.ones([size*2*(R),size*2*(C),3]) * 255
	for i in range(R):
		for j in range(C):
			if i * C + j in indices:
				color = [0,0,255]
			else:
				color = [0,0,0]
			cv2.circle(img, (size*(2*j+1),size*(2*i+1)), radius=size*2//3, color=color, thickness=-1)
	plt.figure(figsize=(10,10))
	# plt.axis('off')
	plt.imshow(img[...,::-1])
	cv2.imwrite(filename, img)
	print('%.2f' % (1-1.*len(indices)/(R*C)))

round0 = [1,14,18,59,71,72,77,85,91,93]
make_image(R=5, C=20, indices=round0, filename='../outputs/group_testing_round-0.png')

R = len(round0)
C = 15
round1 = [  4,   9,  13,  39,  59,  60,  63,  66,  68,  69,  76,  78,  79,
        81,  82,  84,  85,  91,  92, 100, 103, 104, 105, 108, 111, 117,
       120, 124, 125, 126, 134, 135, 138, 144, 147]
make_image(R=R, C=C, indices=round1, filename='../outputs/group_testing_round-1.png')
make_image(R=R, C=1, indices=np.arange(R), filename='../outputs/group_testing_round-1_column.png')

round1B = [9,76,79]
make_image(R=R, C=C, indices=round1B, filename='../outputs/group_testing_round-1B.png')
R = len(round1B)
C = 15
round2 = [4, 7, 9, 13, 16, 25, 28, 30, 33, 36, 37]
make_image(R=R, C=C, indices=round2, filename='../outputs/group_testing_round-2.png')
make_image(R=R, C=1, indices=np.arange(R), filename='../outputs/group_testing_round-2_column.png')

```
